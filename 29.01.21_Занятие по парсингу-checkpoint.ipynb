```python
import matplotlib.pyplot as plt
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response = requests.get("http://online.igk-group.ru/ru/home?name=&ogrn=&inn=6311072190")
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
td = soup.select('tr')
```

```python
data = []
for i in td:
    a = i.get_text().replace("\r\n", "").replace("\n", "").replace("\t", "").replace("  ", "")
    data.append(a)
    
data
```
