### Сглаживаем данные за счет усреднения температуры воздуха понедельно

```python
df_week = df_full.resample('W').mean()
df_week
```

```python
fig, ax = plt.subplots(2,1, figsize=(17,8))

ax[0].plot(df_week['meantemp'])
ax[1].plot(df_week['meantemp'].loc["2015":"2016"])

fig = plt.figure(figsize=(20,5))
plt.show();
```

##### График стал более сглаженным

```python
plt.figure(figsize=(20,7))
plt.rcParams['figure.figsize'] = [20, 10]
sm.tsa.seasonal_decompose(df_week.meantemp).plot()
#plt.yticks(range(0, 4), ('', '', '', ''))
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_week.meantemp)[1])
plt.show();
```

```python
fig,ax = plt.subplots(2,1,figsize=(20,10))
fig = sm.graphics.tsa.plot_acf(df_week.meantemp, lags=52, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(df_week.meantemp, lags=52, ax=ax[1])
plt.show()
```

```python
df_week['meantemp_box'], lmbda = stats.boxcox(df_week.meantemp)
plt.figure(figsize=(20,5))
plt.plot(df_week.meantemp_box)
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
plt.figure(figsize=(20,7))
plt.rcParams['figure.figsize'] = [20, 10]
sm.tsa.seasonal_decompose(df_week.meantemp_box).plot()
#plt.yticks(range(0, 4), ('', '', '', ''))
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_week.meantemp_box)[1])
plt.show();
```

```python
plt.plot(df_week.meantemp_box.diff().dropna())
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
plt.figure(figsize=(20,4))
plt.rcParams['figure.figsize'] = [20, 10]
sm.tsa.seasonal_decompose(df_week.meantemp_box.diff().dropna()).plot()
#plt.yticks(range(0, 4), ('', '', '', ''))
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_week.meantemp_box.diff().dropna())[1])
plt.show();
```

```python
fig,ax = plt.subplots(2,1,figsize=(20,10))
fig = sm.graphics.tsa.plot_acf(df_week.meantemp_box.diff().dropna(), lags=52, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(df_week.meantemp_box.diff().dropna(), lags=52, ax=ax[1])
plt.show()
```

```python
'''
(p,d,q)(P,D,Q)s

p - (0,5)
d - 1 + (0,2)
q - (0,5)

P - (0, 2)
D - 0 + (0,2)
Q - 1 + (0,2)
s - 52
'''
```

```python
from pmdarima.arima import auto_arima

step_wise=auto_arima(df_week.meantemp_box.loc[:"2016"], 
 start_p=5, start_q=0, 
 max_p=6, max_q=6, start_P=0, start_Q=0, 
 max_P=2, max_Q=2, seasonal=True, 
 d=1, max_d=3, m=52, D=1, max_D = 3,
 trace=True, 
 error_action='ignore',
 suppress_warnings=True, 
 stepwise=True)
```

```python
arima3 = model3=sm.tsa.statespace.SARIMAX(df_week.meantemp_box.loc[:"2016"],
                                order=(5,1,1),
                                seasonal_order=(2,1,0, 52), freq='W'
                                #enforce_stationarity=False, enforce_invertibility=False
                                       ).fit(disp=-1)
res = arima3.resid

fig,ax = plt.subplots(2,1,figsize=(15,8))
fig = sm.graphics.tsa.plot_acf(res, lags=50, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(res, lags=50, ax=ax[1])
plt.show()

print(arima3.summary())
```

```python
date_list = df_week.loc['2017-01-01':'2017-04-30'].index

df_day_predict = pd.DataFrame({'timestamp': date_list, 'model': 0})
df_day_predict=df_day_predict.set_index('timestamp')
df_day_predict.index = pd.to_datetime(df_day_predict.index, format="%Y-%m-%d")
df_day_predict['model'] = inv_boxcox(arima3.predict(start='2017-01-01', end='2017-04-30'), lmbda)

plt.figure(figsize=(15,5))
plt.style.use('seaborn')

plt.plot(df_day_predict.model, sns.xkcd_rgb["pale red"])
plt.plot(df_week.meantemp, sns.xkcd_rgb["denim blue"], alpha=0.6)
plt.legend(labels = ['Предсказание SARIMAX', 'Реальные показания'])
plt.xticks(rotation = 45, ha="right")
plt.tight_layout()
```

### Данная модель может использоваться для предсказания погоды на несколько месяцев вперед.
